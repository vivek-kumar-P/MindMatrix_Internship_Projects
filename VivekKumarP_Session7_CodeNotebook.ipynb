## Cell 2 — calculateTip function

```kotlin
import kotlin.math.ceil

// Core tip calculation logic
// roundUp = true -> ceil() rounds up to next whole number
fun calculateTip(amount: Double, tipPercent: Int, roundUp: Boolean): Double {
    val tip = amount * tipPercent / 100.0
    return if (roundUp) ceil(tip) else tip
}

println(calculateTip(100.0, 15, false))  // 15.0
println(calculateTip(100.0, 15, true))   // 15.0 (already whole)
println(calculateTip(75.20, 18, false))  // 13.536
println(calculateTip(75.20, 18, true))   // 14.0 (rounded up)
```

## Cell 3 — formatAmount function

```kotlin
import java.util.Locale

// Formats Double to 2 decimal places
// Locale.US ensures dot as decimal separator everywhere
fun formatAmount(amount: Double): String {
    return String.format(Locale.US, "%.2f", amount)
}

println(formatAmount(7.5))      // 7.50
println(formatAmount(13.536))   // 13.54
println(formatAmount(100.0))    // 100.00
```

## Cell 5 — Input Validation Logic

```kotlin
// Validates all inputs before allowing calculation
// Returns error message string or null if valid

fun validateBill(input: String): String? {
    val value = input.toDoubleOrNull()
    return when {
        input.isBlank() -> "Bill amount cannot be empty"
        value == null   -> "Bill amount must be a number"
        value <= 0      -> "Bill amount must be greater than 0"
        else            -> null   // valid
    }
}

fun validatePeople(input: String): String? {
    val value = input.toIntOrNull()
    return when {
        input.isBlank() -> "Number of people cannot be empty"
        value == null   -> "Must be a whole number"
        value < 1       -> "At least 1 person required"
        else            -> null   // valid
    }
}

fun validateTip(input: String): String? {
    val value = input.toIntOrNull()
    return when {
        input.isBlank()  -> "Tip percent cannot be empty"
        value == null    -> "Must be a number"
        value !in 0..100 -> "Tip must be between 0 and 100"
        else             -> null   // valid
    }
}

// Test validations
println(validateBill(""))       // Bill amount cannot be empty
println(validateBill("-10"))    // Bill amount must be greater than 0
println(validateBill("100"))    // null (valid)
println("---")
println(validatePeople("0"))    // At least 1 person required
println(validatePeople("abc"))  // Must be a whole number
println(validatePeople("4"))    // null (valid)
println("---")
println(validateTip("150"))     // Tip must be between 0 and 100
println(validateTip("20"))      // null (valid)
```

## Cell 6 — Expense Splitter (Task 2 Feature)

```kotlin
import java.util.Locale
import kotlin.math.ceil

fun formatAmount(amount: Double): String =
    String.format(Locale.US, "%.2f", amount)

// Splits bill among people including tip
// Returns Pair(perPerson, totalTip)
fun splitBill(
    bill: Double,
    people: Int,
    tipPercent: Int,
    roundUp: Boolean
): Pair<Double, Double> {
    val tip = bill * tipPercent / 100.0
    val total = bill + tip
    val perPerson = total / people
    val finalPerPerson = if (roundUp) ceil(perPerson) else perPerson
    return Pair(finalPerPerson, tip)
}

// Scenario 1 - no round up
val (perPerson1, totalTip1) = splitBill(100.0, 4, 20, false)
println("Per person : ${formatAmount(perPerson1)}")   // 30.00
println("Total tip  : ${formatAmount(totalTip1)}")    // 20.00
println("---")

// Scenario 2 - with round up
val (perPerson2, totalTip2) = splitBill(75.20, 3, 18, true)
println("Per person : ${formatAmount(perPerson2)}")   // 30.00 (ceiled)
println("Total tip  : ${formatAmount(totalTip2)}")    // 13.54
```

## Cell 7 — History / Save Logic (Task 3 Feature)

```kotlin
import java.util.Locale

fun formatAmount(amount: Double): String =
    String.format(Locale.US, "%.2f", amount)

// Data class to store each saved split
data class SplitRecord(
    val bill: Double,
    val people: Int,
    val tipPercent: Int,
    val perPerson: Double,
    val roundedUp: Boolean
)

// Start with empty history
var historyList: List<SplitRecord> = emptyList()

// Simulate saving 3 records - immutable update each time
historyList = historyList + SplitRecord(100.0, 4, 20, 30.00, false)
historyList = historyList + SplitRecord(75.20, 3, 18, 30.00, true)
historyList = historyList + SplitRecord(200.0, 5, 15, 46.00, false)

// Print history
println("--- Split History ---")
historyList.forEachIndexed { i, record ->
    println(
        "${i + 1}. Bill: ${formatAmount(record.bill)} | " +
        "People: ${record.people} | " +
        "Tip: ${record.tipPercent}% | " +
        "Per person: ${formatAmount(record.perPerson)}"
    )
}
println("Total records: ${historyList.size}")

// Clear history
historyList = emptyList()
println("After clear: ${historyList.size} records")

// OUTPUT:
// --- Split History ---
// 1. Bill: 100.00 | People: 4 | Tip: 20% | Per person: 30.00
// 2. Bill: 75.20  | People: 3 | Tip: 18% | Per person: 30.00
// 3. Bill: 200.00 | People: 5 | Tip: 15% | Per person: 46.00
// Total records: 3
// After clear: 0 records
```
